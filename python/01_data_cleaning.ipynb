```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/raw/DataCoSupplyChainDataset.csv', encoding='latin-1')

print("Shape:", df.shape)
print("\nColumn names:")
for col in df.columns:
    print(f"  {col}")
```

```python
# Select only the columns we need
cols_needed = [
    'Order Id',
    'order date (DateOrders)',
    'shipping date (DateOrders)',
    'Days for shipping (real)',
    'Days for shipment (scheduled)',
    'Late_delivery_risk',
    'Delivery Status',
    'Customer Id',
    'Customer Segment',
    'Market',
    'Order Region',
    'Order Country',
    'Order City',
    'Category Name',
    'Product Name',
    'Sales',
    'Order Profit Per Order',
    'Order Item Discount',
    'Order Item Quantity',
    'Shipping Mode'
]

df = df[cols_needed]

# Rename to snake_case
df.columns = [
    'order_id', 'order_date', 'ship_date',
    'days_for_shipping_real', 'days_for_shipment_scheduled',
    'late_delivery_risk', 'delivery_status',
    'customer_id', 'customer_segment',
    'market', 'order_region', 'order_country', 'order_city',
    'category_name', 'product_name',
    'sales', 'order_profit_per_order',
    'order_item_discount', 'order_item_quantity',
    'shipping_mode'
]

print("Shape after selecting columns:", df.shape)
print("\nNull counts:")
print(df.isnull().sum())
```

```python
# Fix date columns
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
df['ship_date'] = pd.to_datetime(df['ship_date'], errors='coerce')

# Add time columns
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.to_period('M').astype(str)

# Check delivery status and shipping mode values
print("Delivery Status values:")
print(df['delivery_status'].value_counts())

print("\nShipping Mode values:")
print(df['shipping_mode'].value_counts())

print("\nOrder Year values:")
print(df['order_year'].value_counts().sort_index())

print("\nShape:", df.shape)
```

```python
# Save cleaned file
df.to_csv('../data/cleaned/supply_chain_cleaned.csv', index=False)

print("Saved successfully.")
print("Shape:", df.shape)
print("Columns:", list(df.columns))
```
